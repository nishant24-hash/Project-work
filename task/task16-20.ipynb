```python
# Task 16: Display strings in lowercase alphabetical order

strings = []

while True:
    s = input("Enter a string: ")
    if s == "":
        break
    strings.append(s.lower())

for item in sorted(strings):
    print(item)
```

```python
# Task 17: Count vowels in a string

text = input("Enter a string: ")
vowels = "aeiouAEIOU"

count = 0
for char in text:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```

```python
# Task 18: Check if all numbers are between 0 and 100

def valid_scores(numbers):
    return all(0 <= n <= 100 for n in numbers)

# Demonstration
print(valid_scores([10, 50, 100]))   # True
print(valid_scores([10, 150]))       # False
```

```python
# Task 19: Check if mark is a pass

mark = int(input("Enter exam mark: "))

if mark >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
# Task 20: Check if list contains no empty strings

def no_empty_strings(words):
    return "" not in words

# Demonstration
print(no_empty_strings(["apple", "banana"]))   # True
print(no_empty_strings(["apple", ""]))         # False
```
