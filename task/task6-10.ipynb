```python
# Task 6: Count even numbers entered by user

count = 0

while True:
    value = input("Enter an integer: ")
    if value == "":
        break
    if int(value) % 2 == 0:
        count += 1

print("Number of even numbers:", count)
```

```python
# Task 7: Check if string is a palindrome

text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Task 8: Find maximum value in a list

def find_max(numbers):
    return max(numbers)

# Demonstration
print(find_max([3, 7, 2, 9]))  # 9
```

```python
# Task 9: Check divisibility by 3 and 7

num = int(input("Enter an integer: "))

if num % 3 == 0 and num % 7 == 0:
    print("Divisible by both 3 and 7")
else:
    print("Not divisible by both")
```

```python
# Task 10: Count unique strings

strings = []

while True:
    s = input("Enter a string: ")
    if s == "":
        break
    strings.append(s)

print("Unique strings count:", len(set(strings)))
```
