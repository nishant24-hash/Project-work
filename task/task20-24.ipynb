```python
# Task 21: Display sum of entered integers

total = 0

while True:
    value = input("Enter an integer: ")
    if value == "":
        break
    total += int(value)

print("Sum:", total)
```

```python
# Task 22: Check if string is a palindrome

text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Task 23: Calculate average of a list

def average(numbers):
    return sum(numbers) / len(numbers)

# Demonstration
print(average([10, 20, 30]))  # 20.0
```

```python
# Task 24: Average length of strings

strings = []

while True:
    s = input("Enter a string: ")
    if s == "":
        break
    strings.append(s)

average_length = sum(len(s) for s in strings) / len(strings)
print("Average length:", average_length)
```
