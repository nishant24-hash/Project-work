```python
# Task 11: Check if number is odd or even

num = int(input("Enter an integer: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Task 12: Check if list is sorted in ascending order

def is_sorted(numbers):
    return numbers == sorted(numbers)

# Demonstration
print(is_sorted([1, 2, 3]))     # True
print(is_sorted([3, 1, 2]))     # False
```

```python
# Task 13: Display the longest string entered

strings = []

while True:
    s = input("Enter a string: ")
    if s == "":
        break
    strings.append(s)

print("Longest string:", max(strings, key=len))
```

```python
# Task 14: Check if number is a multiple of 5

num = int(input("Enter an integer: "))

if num % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

```python
# Task 15: Check if all strings start with a capital letter

def all_capital(words):
    return all(word[0].isupper() for word in words)

# Demonstration
print(all_capital(["Apple", "Banana"]))   # True
print(all_capital(["Apple", "banana"]))   # False
```
