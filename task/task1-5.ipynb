```python
# Task 1: Write a function that checks if all numbers in a list are unique

def all_unique(numbers):
    return len(numbers) == len(set(numbers))

# Code to demonstrate correctness
print(all_unique([1, 2, 3, 4]))   # True
print(all_unique([1, 2, 2, 4]))   # False
```

```python
# Task 2: Check if runner time is below the standard time

time = float(input("Enter lap time in seconds: "))

if time < 50.5:
    print("Below standard time")
else:
    print("Not below standard time")
```

```python
# Task 3: Display entered strings in reverse order

strings = []

while True:
    s = input("Enter a string: ")
    if s == "":
        break
    strings.append(s)

for item in reversed(strings):
    print(item)
```

```python
# Task 4: Check if number is positive, negative, or zero

num = int(input("Enter an integer: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Task 5: Check if list contains "coconut"

def contains_coconut(words):
    return "coconut" in words

# Demonstration
print(contains_coconut(["apple", "banana", "coconut"]))  # True
print(contains_coconut(["apple", "banana"]))             # False
```
