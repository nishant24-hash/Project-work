```python
# Initialize counters for vowels, consonants, digits and special characters
vowels = consonants = digits = special = 0

# Open the file in read mode
with open("input.txt", "r") as file:
    text = file.read()  # Read entire content of file

# Loop through each character in text
for ch in text:
    if ch.lower() in "aeiou":  # Check if character is vowel
        vowels += 1
    elif ch.isalpha():  # Check if character is letter (but not vowel)
        consonants += 1
    elif ch.isdigit():  # Check if character is digit
        digits += 1
    else:  # Anything else is special character
        special += 1

# Print counts
print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special Characters:", special)
```

```python
# Open the first and second file in read mode
with open("file1.txt", "r") as f1, open("file2.txt", "r") as f2:
    content1 = f1.read()  # Read first file
    content2 = f2.read()  # Read second file

# Open a third file in write mode to store merged content
with open("merged.txt", "w") as f3:
    f3.write(content1)  # Write content of first file
    f3.write("\n----- End of File 1 -----\n")  # Add separator
    f3.write(content2)  # Write content of second file

print("Files merged successfully.")
```

```python
import csv

# Function to calculate grade based on marks
def grade(mark):
    if mark >= 80:
        return "A"
    elif mark >= 60:
        return "B"
    else:
        return "C"

# Open CSV file containing student names and marks
with open("students.csv", "r") as file:
    reader = csv.reader(file)
    print("Name\tMark\tGrade")
    for row in reader:
        name = row[0]  # First column is name
        mark = int(row[1])  # Second column is mark
        print(name, mark, grade(mark), sep="\t")  # Print formatted table
```

```python
from collections import Counter

# Open the file and read content
with open("input.txt", "r") as file:
    words = file.read().lower().split()  # Convert to lowercase and split words

# Count frequency of each word
count = Counter(words)
most_common = count.most_common(1)[0]  # Get most common word

print("Most frequent word:", most_common[0])
print("Occurrences:", most_common[1])
```

```python
# Read file content
with open("input.txt", "r") as file:
    words = set(file.read().lower().split())  # Convert to set to remove duplicates

sorted_words = sorted(words)  # Sort alphabetically

# Write unique sorted words into a new file
with open("unique_words.txt", "w") as file:
    for word in sorted_words:
        file.write(word + "\n")
```

```python
count = 0

# Open file to read lines
with open("input.txt", "r") as file:
    for line in file:
        if line.strip().lower().startswith(('a','e','i','o','u')):  
            # strip() removes leading/trailing spaces
            count += 1

print("Lines starting with vowel:", count)
```

```python
# Open input file in read mode and output file in write mode
with open("input.txt", "r") as file, open("reversed.txt", "w") as out:
    for line in file:
        out.write(line[::-1])  # Reverse each line using slicing
```

```python
import csv

dept_salary = {}  # Dictionary to store salaries per department

# Open CSV file containing employee data
with open("employees.csv", "r") as file:
    reader = csv.reader(file)
    for name, dept, salary in reader:
        salary = float(salary)
        # Add salary to the list of the department
        dept_salary.setdefault(dept, []).append(salary)

# Calculate and print average salary for each department
for dept, salaries in dept_salary.items():
    avg = sum(salaries) / len(salaries)
    print(dept, "Average Salary:", avg)
```

```python
# Define Student class with required attributes
class Student:
    def __init__(self, id, name, address, year, level, section):
        self.id = id
        self.name = name
        self.address = address
        self.year = year
        self.level = level
        self.section = section

    # Method to display student info
    def display(self):
        print(self.id, self.name, self.address,
              self.year, self.level, self.section)

# Create a student object and display
s = Student(1, "Nishant", "Kathmandu", 2024, "Bachelor", "A")
s.display()
```

```python
import csv

# Define Employee class with required attributes
class Employee:
    def __init__(self, empid, name, address, contact, spouse, children, salary):
        self.empid = empid
        self.name = name
        self.address = address
        self.contact = contact
        self.spouse = spouse
        self.children = children
        self.salary = salary

try:
    # Open employees.csv in append mode
    with open("employees.csv", "a", newline="") as file:
        writer = csv.writer(file)
        # Create an employee object
        emp = Employee(1, "Ram", "Pokhara", "9800000000", "Sita", 2, 50000)
        # Write employee details to file
        writer.writerow([emp.empid, emp.name, emp.address,
                         emp.contact, emp.spouse, emp.children, emp.salary])
except Exception as e:
    print("Error:", e)  # Handle any file errors
```

```python
# Define Library class with book management methods
class Library:
    def issue_book(self, book):
        with open("library.txt", "a") as file:
            file.write(book + " - Issued\n")  # Mark book as issued

    def return_book(self, book):
        with open("library.txt", "a") as file:
            file.write(book + " - Returned\n")  # Mark book as returned

    def search_book(self, book):
        try:
            with open("library.txt", "r") as file:
                for line in file:
                    if book in line:  # Check if book exists in file
                        print("Book found:", line)
        except FileNotFoundError:
            print("Library file not found")  # Handle missing file

# Create library object and test methods
lib = Library()
lib.issue_book("Python Programming")
lib.search_book("Python Programming")
```
